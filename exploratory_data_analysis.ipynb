```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Concrete_Data.csv')
```

```python
df.columns = ['cement', 'BFS', 'FA', 'water', 'SPZ', 'coarse', 'fine', 'age', 'strength']
```

```python
df['W/C'] = df['water'] / df['cement']
df['A/C'] = (df['fine'] + df['coarse']) / df['cement']
df['F/C'] = df['fine'] / df['cement']
df['C/C'] = df['coarse'] / df['cement']
df.round(decimals=2)
```

```python
df['W/C'] = df['water'] / df['cement']
df['C/A'] = df['cement'] / (df['fine'] + df['coarse']) 
df['C/F'] = df['cement'] / df['fine']
df['C/C'] = df['cement'] / df['coarse']
df.round(decimals=2)

print(max(df['W/C']))
print(max(df['C/A']))
print(max(df['C/F']))
print(max(df['C/C']))
df
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data=df.corr(), annot=True)
```

```python
def plot(col1: str, filename):
    
    x = df[col1]
    y = df['strength']
    
    sns.set_style('whitegrid')
    sns.regplot(x=x, y=y)
    
#     plt.savefig(f'C:\\Users\\boate\\OneDrive\\Desktop\\JOURNAL-PAPER\\LATEX\\Elsevier Article (elsarticle) Template\\{filename}.jpg', bbox_inches='tight', dpi=100)
```

```python
plot(col1='W/C', filename='water_cement')
```
